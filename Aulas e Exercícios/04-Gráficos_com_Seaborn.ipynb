Importação

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

Importando Base de Dados

```python
df_videos = pd.read_excel("Arquivos/videosYT.xlsx", "videos")
display(df_videos)
```

Criando um gráfico de dispersão:

Documentação : https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
fig = sns.scatterplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue= 'Responsável', style= 'Responsável', palette= ['red', 'black'])
plt.show(fig)
```

Criando um gráfico de dispersão (relacional):


Documentação : https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
gráfico_rel = sns.relplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue= "Categoria", col="Responsável")
gráfico_rel.set_titles("O Responsável deste gráfico é o {col_name}")
plt.show(gráfico_rel)
```

Criando um gráfico de linha:

Documentação : https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot

```python
df_inscritos = pd.read_excel("Arquivos/videosYT.xlsx", "Inscritos")

graf_linha = sns.lineplot(data=df_inscritos, x = "Mês/Ano", y= "Inscritos", color = 'red')
plt.show(graf_linha)
```

Criando Histogramas:

```python
fig = sns.displot(data=df_videos, x='Nº de Views', hue= "Responsável", col= "Responsável")
plt.show(fig)
```

Criando uma regressão linear:


Documentação: https://seaborn.pydata.org/tutorial/regression.html#functions-to-draw-linear-regression-models

```python
graf_regre = sns.regplot(data=df_videos, x='Nº de Views', y='Nº de Likes')
```

```python
graf_regre = sns.lmplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue= "Responsável", markers=['o', 'x'])
plt.show(graf_regre)
```
